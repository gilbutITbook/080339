```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline

plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False

from IPython.display import set_matplotlib_formats
set_matplotlib_formats('retina')
```

```python
#    

filepath = './data'
population = pd.read_csv(filepath + "/__2022_05.csv", encoding='euc-kr')
population.head(2)
```

```python
print("  :", population.shape)
```

```python
# 
population.isnull().sum()
```

```python
#        .  
popula_female = population.iloc[:,-12:]
popula_info = population.iloc[:,:4]
pop_female_info = pd.concat([popula_info, popula_female], axis=1)
```

```python
pop_female_info.head(3)
```

```python
# 2022  
pop_female_info[''].value_counts().sort_index()
```

```python
pop_female_info = pop_female_info[pop_female_info['']!=2022]
```

```python
pop_female_info.rename(columns={"0~9 ()":"10_",
                                "10~19 ()":"10",
                                "20~29 ()":"20",
                                "30~39 ()":"30",
                                "40~49 ()":"40",
                                "50~59 ()":"50",
                                "60~69 ()":"60"
                                }, inplace=True)
```

```python
pop_female_info['70_']=pop_female_info["70~79 ()"]+pop_female_info["80~89 ()"]+\
pop_female_info["90~99 ()"] + pop_female_info["100  ()"]
```

```python
pop_female_info.info()
```

```python
df_female = pop_female_info.drop(['70~79 ()', '80~89 ()', '90~99 ()','100  ()'], axis=1)
```

```python
df_female.head(2)
```

```python
"""    """
df_female_dong=df_female[df_female[''].str.contains('')]
df_female_dong.head()
```

```python
df_female_dong.shape
```

```python
# 40   (%) 50  (%) 
under40s = df_female_dong['10_']+df_female_dong['10']+df_female_dong['20']+df_female_dong['30']+df_female_dong['40']
over50s = df_female_dong['50']+df_female_dong['60']+df_female_dong['70_']

under40_percent = round((under40s / df_female_dong['  ()']) * 100, 2)
over50_percent = round((over50s / df_female_dong['  ()']) * 100, 2)

df_female_dong['40__(%)'] = under40_percent
df_female_dong['50__(%)'] = over50_percent

df_female_dong.head(2)
```

```python
# split() 
x = "   "
print("split()   :",x)
print("split()   :",x.split())
print("   :",x.split()[-1])
print("   :",x.split()[1])
```

```python
df_female_dong[''] = df_female_dong[''].apply(lambda x:x.split()[-1])
df_female_dong[''] = df_female_dong[''].apply(lambda x:x.split()[1])
```

```python
df_female_dong.sample()
```

```python
df_female_dong['-'] = df_female_dong['']+ " " +df_female_dong['']
```

```python
df_result = df_female_dong[['','','','-','  ()','40__(%)', '50__(%)']]
```

```python
df_result.isnull().sum()
```

```python
#    
df_result[df_result['40__(%)'].isnull()]
```

```python
df_result.dropna(inplace=True)
```

```python
#  5   (2016 ~ 2021)
df_resent = df_result[df_result['']>2015]
```

```python
df_result.sample(3)
```

```python
df_resent.sample()
```

```python
df_resent_group = df_resent.groupby(['-'])['40__(%)', "50__(%)"].mean()
```

```python
df_resent_group.reset_index(inplace=True)
```

```python
df_resent_group.sort_values(by='40__(%)', ascending=False)[:20]
```

```python
df_top20 = df_resent_group.sort_values(by='40__(%)', ascending=False)[:20]
```

```python
x = df_top20['-']
y = df_top20['40__(%)']
```

```python
plt.figure(figsize=(10, 5))
plt.title("2016 ~ 2021, 40  \n     20 ", fontsize=15)
plt.bar(x, y, alpha=0.7, color='darkgreen')
plt.xticks(size=12, rotation=45)
plt.yticks(size=12)
plt.ylim(70,90)
plt.ylabel('%', size=20)
plt.grid(True, alpha=0.2)
```

```python
plt.figure(figsize=(8, 8))
plt.title("2016 ~ 2021, 40  \n     20 ", fontsize=20)
plt.barh(x, y, alpha=0.7, color='darkgreen')
plt.xticks(size=15)
plt.yticks(size=15)
plt.xlim(50,90)
plt.grid(True, alpha=0.2)
```
